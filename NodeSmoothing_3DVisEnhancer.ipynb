```python
import pandas as pd

file_path = '/content/node_positions_basillus_42C45m.csv'

# Read the CSV data into a DataFrame
df = pd.read_csv(file_path)

# Display the first few rows of the DataFrame to verify
print(df.head())
```

```python
  ##  1: range(1, 285),
  ##  2: range(285, 492),
  ##  3: range(492, 546),
  ##  4: range (546, 569)
def assign_ChID(id_str):
    # Extract the numeric part from the string, assuming the format is 'nodeX'
    # where X is the number. This strips out 'node' and converts the rest to an integer.
    numeric_part = int(id_str.lstrip('Node'))

    # Apply your conditions based on the numeric part
    if 1 <= numeric_part <= 568:
        return 1
    # Add more conditions here as needed
    elif 569 <= numeric_part <= 1012:
      return 2
    elif 1013 <= numeric_part <= 1258:
      return 3

    else:
        return 0  # Or any other default value for IDs outside your specified ranges

# Apply the function to the 'id' column to create the new 'ChID' column
df['ChID'] = df['id'].apply(assign_ChID)
```

```python
def smooth_positions(df, window_size=5):
    # Initialize columns for smoothed coordinates
    df_smoothed = df.copy()
    df_smoothed['x_smoothed'] = 0
    df_smoothed['y_smoothed'] = 0
    df_smoothed['z_smoothed'] = 0

    half_window = window_size // 2

    # Iterate through the DataFrame to calculate smoothed coordinates
    for i in range(len(df)):
        # Determine the window range; handle edge cases by limiting the range to valid indices
        start = max(i - half_window, 0)
        end = min(i + half_window + 1, len(df))

        # Calculate the mean for x, y, and z within the window
        df_smoothed.loc[i, 'x_smoothed'] = df['x'][start:end].mean()
        df_smoothed.loc[i, 'y_smoothed'] = df['y'][start:end].mean()
        df_smoothed.loc[i, 'z_smoothed'] = df['z'][start:end].mean()

    return df_smoothed
```

```python
# Number of iterations for smoothing
num_iterations = 5

for _ in range(num_iterations):
    # Apply the smoothing function
    df = smooth_positions(df)
    # Update the original coordinate columns with the smoothed coordinates
    df['x'] = df['x_smoothed']
    df['y'] = df['y_smoothed']
    df['z'] = df['z_smoothed']

# Display the first few rows after smoothing to verify
print(df.head())
```

```python
fig = px.scatter_3d(df, x='x_smoothed', y='y_smoothed', z='z_smoothed',
                    title="3D Scatter Plot of Nodes",
                    labels={'x': 'X Axis', 'y': 'Y Axis', 'z': 'Z Axis'})

fig.update_traces(marker=dict(size=2))  # Adjust the size as needed

# Show the figure
fig.show()
```
